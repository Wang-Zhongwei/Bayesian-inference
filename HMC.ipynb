```python
import numpy as np
import matplotlib.pyplot as plt

'''Leap frog method'''
def leapfrog(A, u, N):
    orbit = np.zeros((N,2))

    dt = 2*np.pi/N
    for i in range(N):
        u[1] = u[1] + dt/2 * A[1] @ u
        u[0] = u[0] + dt * A[0] @ u
        u[1] = u[1] + dt/2 * A[1] @ u
        orbit[i] = u
    return orbit
```

```python
'''
A[0] = partial H / partial p. 
A[1] = -partial H / partial q.
Linear system is assumed so that partial H / partial q is a linear combination of
q.
'''
A = np.array([[0,1],[-1,0]])
u = np.array([1.0,0.0]) # initial condition
N = 64

orbit = leapfrog(A, u, N)
```

```python
# phase space plot
plt.plot(orbit[:,0], orbit[:,1])
plt.show()

# energy plot
plt.plot([p @ p for p in orbit])
plt.show()
```

```python
'''Trapezoidal rule'''
import scipy.linalg as la

def trapezoidal(A, u, N):
    p = len(u)
    orbit = np.zeros((N,p))

    dt = 2*np.pi/N
    for i in range(N):
        u = la.inv(np.eye(p) - dt/2 * A) @ (np.eye(p) + dt/2 * A) @ u
        orbit[i] = u
    return orbit
```

```python
A = np.array([[0,1],[-1,0]])
u = np.array([1.0,0.0])
N = 64
orbit = trapezoidal(A, u, N)
```

```python
# phase space plot
plt.plot(orbit[:,0], orbit[:,1])
plt.show()

# energy plot
plt.plot([p @ p for p in orbit])
plt.show()
```
